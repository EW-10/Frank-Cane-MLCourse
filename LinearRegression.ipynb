# Linear Regression

Let's fabricate some data that shows a roughly linear relationship between page speed and amount purchased:

```python
# Importing libraries
%matplotlib inline
import numpy as np
from pylab import *

# Creating data
pageSpeeds = np.random.normal(3.0, 2.0, 1000)
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 10, 1000)) * 3

# Clipping the data to make sure there are no negative values
purchaseAmount = purchaseAmount.clip(min = 0)
pageSpeeds = pageSpeeds.clip(min = 0.2)

# Creating a plot
scatter(pageSpeeds, purchaseAmount, c = 'r', edgecolor = 'blue', s = 80)

# Displaying the first 3 values of the dataset
display(pageSpeeds[:3], purchaseAmount[:3])

# Adding titles and labels
plt.xlabel('Page Rendering Speed (Sec)', c = 'b', fontsize = 16)
plt.ylabel('Purchase Amount ($)', c = 'b', fontsize = 16)
plt.title('Purchase Amount vs Page Rendering Speed', fontsize = 18, c = 'g', weight = 'bold')
plt.savefig('C:\\MLCourse\\myplot10.png', format = 'png')
```

As we only have two features, we can keep it simple and just use scipy.state.linregress:

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)
```

Not surprisngly, our R-squared value shows a really good fit:

```python
r_value ** 2
```

Let's use the slope and intercept we got from the regression to plot predicted values vs. observed:

```python
import matplotlib.pyplot as plt

def predict(x):
    return slope * x + intercept

fitLine = predict(pageSpeeds)

plt.scatter(pageSpeeds, purchaseAmount, c = 'g')
plt.plot(pageSpeeds, fitLine, c='b', lw = 6)
plt.show()
```
